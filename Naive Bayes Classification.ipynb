# Gaussian Naive Bayes

1. import a make_blobs function from sklearn's datasets
2. create a set of data using make_blobs which has 100 samples, 2 labels, random_states as 2, and cluster standard deviation as 1.5.
3. import GaussianNB estimator from sklearn's naive_bayes
4. fit
5. create a test sample set with using random module in numpy
6. plot both the training set and the testing set


```python
from sklearn.datasets import make_blobs
X,y = make_blobs(100, 2, centers = 2, random_state = 2, cluster_std = 1.5)
plt.scatter(X[:,0], X[:,1],c = y, s = 50, cmap = 'RdBu')
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X,y)
```

```python
rng = np.random.RandomState(0)
mu = np.empty(2, dtype = np.ndarray)
std= np.empty(2, dtype = np.ndarray)
for label, color in enumerate(['red', 'blue']):
    mask = (y == label)
    mu[label], std[label] = X[mask].mean(0), X[mask].std(0)
```

```python
mu, std
```

```python
Xnew = [-6, -14] + [14, 18] * rng.rand(2000, 2)
ynew = model.predict(Xnew)
```

```python
plt.scatter(X[:, 0], X[:,1], c = y, s = 50, cmap = 'RdBu')
lim = plt.axis()
plt.scatter(Xnew[:,0], Xnew[:,1], c = ynew, s = 20, cmap = 'RdBu', alpha = 0.1)
plt.axis(lim)
```

```python
yprob = model.predict_proba(Xnew)
yprob[-8:].round(2)
```
